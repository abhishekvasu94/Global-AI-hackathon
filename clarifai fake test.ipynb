```python
fake_test = "C:/Users/Roshan Sridhar/Downloads/photoshopbattles FAKE TEST/*"
count = 0
model=app.models.get('fake_images')
for f in glob.glob(fake_test):
    count += 1
    try:
        image = app.inputs.create_image_from_filename(f)
        url=model.predict([image])['outputs'][0]['input']['data']['image']['url']
        pred=model.predict([image])['outputs'][0]['data']['concepts'][0]['value']
        if (model.predict([image])['outputs'][0]['data']['concepts'][0]['name'])=='real':
            test_df  = test_df.append({"url":str(url),"score":0,"real":pred},ignore_index=True)
        else:
            test_df  = test_df.append({"url":str(url),"score":0,"fake":pred},ignore_index=True)
    except:
        continue
        
    if count%25==0:
        print(count)
    if count>500:
        break
```
